```python
#import data from csv files
headphone = pd.read_csv("data/Headphone Data.csv",index_col=[0])
laptop = pd.read_csv("data/Laptop Data.csv",index_col=[0])
smart_watch = pd.read_csv("data/Smart Watch.csv",index_col=[0])
tablet = pd.read_csv("data/Tablet and Ebook Data.csv",index_col=[0])
mobile_phone = pd.read_csv("data/Mobile Phone Data.csv",index_col=[0])
```

```python
#insert Product Category for each df
headphone.insert(loc=2, column="Product Category", value="Headphone")
laptop.insert(loc=2, column="Product Category", value="Laptop")
smart_watch.insert(loc=2, column="Product Category", value="Smart Watch")
tablet.insert(loc=2, column="Product Category", value="Tablet")
mobile_phone.insert(loc=2, column="Product Category", value="Mobile Phone")
```

```python
#combine data into 1 dataframe
combined = pd.concat([headphone, laptop, smart_watch, tablet, mobile_phone],ignore_index=True)
combined.head()
```

```python
#function to extract brand name and reassign new column as "Brand" and reassign "Product Name"
def extract_brand(df):
    split_list = df["Product Name"].str.split(" - ", n=1)
    brand_name = [item[0] for item in split_list]
    product_name = [item[1] for item in split_list]
    df.insert(loc=0, column="Brand", value=brand_name)
    df["Product Name"] = product_name
    return df
```

```python
#"Brand" containing mixed English and Chinese
#if "Brand" contains Eng+Chi, extract Eng or reassign element as only Eng

eng_only=[]
for i in combined['Brand']:
    RE = re.compile(u'[⺀-⺙⺛-⻳⼀-⿕々〇〡-〩〸-〺〻㐀-䶵一-鿃豈-鶴侮-頻並-龎]', re.UNICODE)
    nochinese = RE.sub('', i)
    eng_only.append(nochinese)
len(eng_only)
```

```python
combined["Brand"]=eng_only
combined[combined["Brand"].str.contains("ASUS")].head(7)
```

```python
#drop brand called "#N/A"
combined.drop(combined[combined["Brand"]=="#N/A"].index, inplace=True)
```

```python
#re-label brand called "-" as "Others"
combined.loc[combined["Brand"]=="-", "Brand"] = "Others"
```

```python
#Some of brands have "." at the end. For example, there is "JBL" and "JBL."
combined[combined["Brand"]=="JBL."]
```

```python
combined["Brand"] = combined["Brand"].replace("\.$","",regex=True)
```

```python
#Extract the data
combined['Sales Volume'] = combined['Sales Volume'].str.extract('(\d+,?\d+)+')

combined['Number of Review'] = combined['Number of Review'].str.extract('(\d+)')

combined['Price'] = combined['Price'].str.extract('(\d+,?\d+).')
```

```python
#removing the commas and convert to int
combined['Price'] = combined['Price'].str.replace(',', '').astype(int)

combined['Sales Volume'] = combined['Sales Volume'].str.replace(',', '')
combined['Sales Volume']= [int(np.nan_to_num(i)) for i in combined['Sales Volume']]

combined['Number of Review']= [int(np.nan_to_num(i)) for i in combined['Number of Review']]
```

```python
#Rating and convert to floats
combined['Rating'] = combined['Rating'].str.replace('Null', '0')
combined['Rating'] = combined['Rating'].astype(str).astype(float)
```

```python
#function to drop irrelevant products such as cables, chargers, etc.
def irrelevant_product(df_name,col_name): #col_name format: 'col_name'
    exluding_List = ['glasses','case','cable', 'usb','toothbrush','stand','keyboard','vacuum mop','charger']
    exluding_index = []
    for i in range(len(df_name[col_name])):
        for word in exluding_List:
            if word in df_name[col_name][i].lower():
                exluding_index.append(i)
                break
    df = df_name.drop(exluding_index)
    return df
```
