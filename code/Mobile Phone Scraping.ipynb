```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import re
import numpy as np 
import pandas as pd
%matplotlib inline
import seaborn as sns
import matplotlib
matplotlib.rcParams['figure.figsize'] = (30, 20)
```

```python
URL = "https://www.hktvmall.com/hktv/en/main/search?page=0&q=%3Arelevance%3Acategory%3AAA31050000000%3Azone%3Agadgetsandelectronics%3Astreet%3Amain%3A"
driver = webdriver.Chrome("/Users/peteryeboncho/Downloads/chromedriver")
driver.get(URL)
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
wait = driver.implicitly_wait(5)

df=pd.DataFrame(columns= ["Product Name","Sales Volume","Rating","Number of Review"])
df2=pd.DataFrame(columns=["Price"])

while True:
    
    driver.get(driver.current_url)
    subhtml = driver.page_source
    soup = BeautifulSoup(subhtml, "html.parser")
    nextlink = driver.find_element_by_class_name('next-btn')
    ad_find = 'none' not in driver.find_element_by_class_name("CrazyAd").get_attribute('style')
    
    for i in soup.find_all(class_="upper-wrapper"):

        product_name = i.find(class_="brand-product-name").text
        #print(product_name)

        try:
            sales_number=i.find(class_="salesNumber-container").text
        except:
            sales_number= 'Null'
        #print(sales_number)

        try:
            ratings=i.find(class_="star_container")['data-rating']
        except:
            ratings='Null'
        #print(ratings)

        try:
            review_numbers=i.find(class_="review-number").text
        except:
            review_numbers='Null'
        #print(review_numbers)

        df=df.append({"Product Name":product_name, "Sales Volume":sales_number, "Rating":ratings, "Number of Review":review_numbers},ignore_index=True)
        
        
    for i in soup.find_all(class_="lower-wrapper"):

        price=i.find(class_='price').text
        #print(price)

        df2=df2.append({"Price":price},ignore_index=True)
        

    
    if nextlink.get_attribute('class') == 'next-btn':
        if ad_find:
            close_ad = driver.find_element_by_class_name("btnCloseLarge")
            close_ad.click()


        for i in soup.find_all(class_="upper-wrapper"):
            productname = i.find(class_="brand-product-name").text
            print(productname)


        nextlink.click()
        wait
    else:
        print('stop')
        break


```

```python
mobile_phone_data = pd.concat([df,df2],axis=1)
```

```python
mobile_phone_data.to_csv("Mobile Phone Data.csv")
```
